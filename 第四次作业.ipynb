```python
import requests
import json

def get_response():
    url = "https://cn.bing.com/tlookupv3?isVertical=1&&IG=E5735EB3C4FF40969C257FA39910DAF4&IID=translator.5028.4"
    text = input("当前为必应翻译，请宁输入要翻译的内容：")
    data = {
        "from": "zh-Hans",
        "to": "en",
        "text": text
    }
    response = requests.post(url=url, data=data)
    return response.json()[0]


def parser(data):
    result = [item["normalizedTarget"] for item in data["translations"]]
    print(result)


def run():
    data = get_response()
    parser(data)

if __name__ == "__main__":
    run()
```
